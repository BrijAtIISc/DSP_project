```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Companies Data Analysis

```python
article_df = pd.read_csv("./company_news_preprocessed.csv", parse_dates=['publishedAt'])
```

```python
# Verify the count of null values
article_df.isnull().sum()
```

```python
article_df['description'].str.len().plot(kind='hist', bins=20).set_xlabel('description length')
```

```python
article_df['summary'].str.len().plot(kind='hist', bins=20).set_xlabel('BART summary length')
```

```python
plt.figure(figsize=(14, 7))
article_df = article_df.sort_values(by='publishedAt')
plt.plot(article_df['publishedAt'].dt.date, article_df['description'].str.len())
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Description Length')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
article_df = article_df.sort_values(by='publishedAt')
plt.plot(article_df['publishedAt'].dt.date, article_df['description_vader'])
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Description VADER Score')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
article_df = article_df.sort_values(by='publishedAt')
plt.plot(article_df['publishedAt'].dt.date, article_df['summary'].str.len())
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Summary Length')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
article_df = article_df.sort_values(by='publishedAt')
plt.plot(article_df['publishedAt'].dt.date, article_df['summary_vader'])
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Summary VADER Score')
plt.show()
```
